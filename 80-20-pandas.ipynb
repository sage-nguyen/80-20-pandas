```python
df["score_times_10"] = df["Score"] * 10 
```

```python
df.head()
```

```python
def map_score_to_category(score): #created function by user
    if score >= 7:
        return "High Score"
    elif score < 4:
        return "Low Score"
    else:
        return "Medium Score"
```

```python
df["Score"].map(map_score_to_category)
# .apply function: apply the created function for the whole DataFrame; 
# .map function: apply the created function for the Series
```

```python
df["score_category"] = df["Score"].map(map_score_to_category)

                
```
